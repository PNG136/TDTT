#### Câu 1
Yêu cầu. Cước phí Taxi Thông Minh. 

Viết một hàm duy nhất smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính và trả về total (tổng tiền phải trả, 

kiểu số nguyên, đơn vị VND). 
 
Các quy tắc cần cài đặt: 
1. Giá mở cửa (0–0.8 km): - Xe 4 chỗ: 11,000 VND - Xe 7 chỗ: 12,000 VND 
2. Giá theo km sau 0.8 km: - Xe 4 chỗ: từ 0.8→20 km nh 13,000 VND/km; trên 20 km tính 10,500 VND/km - Xe 7 chỗ: từ 0.8→20 km nh 14,000 VND/km; trên 20 km tính 12,000 VND/km 
3. Phí chờ: 5 phút đầu miễn phí; sau đó, mỗi 5 phút (làm tròn lên theo từng block 5 phút) tính 2,000 VND. 
4. Phụ thu ban đêm: nếu start_hour thuộc [22,23,0,1,2,3,4,5], cộng +20% trên tiền tạm tính (mở cửa + quãng đường + 
chờ). 
5. Điều kiện đặc biệt: - Nếu km > 30 và không phải ban đêm, áp dụng giảm 10% trên tổng tiền. - Nếu tổng tiền cuối cùng < 50,000 nhưng km > 5, đặt tổng tiền bằng 50,000 VND (mức tối thiểu).

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    if car_type == 4:
        open_price = 11000
        rate_1 = 13000
        rate_2 = 10500
    else:
        open_price = 12000
        rate_1 = 14000
        rate_2 = 12000

    if km <= 0.8:
        dist_fare = open_price
    elif km <= 20:
        dist_fare = open_price + (km - 0.8) * rate_1
    else:
        dist_fare = open_price + (19.2 * rate_1) + (km - 20) * rate_2

    wait_fare = 0
    if wait_minutes > 5:
        excess_wait = wait_minutes - 5
        blocks = (excess_wait + 4) // 5
        wait_fare = blocks * 2000

    subtotal = dist_fare + wait_fare
    
    is_night = start_hour in [22, 23, 0, 1, 2, 3, 4, 5]
    if is_night:
        subtotal *= 1.2
    elif km > 30:
        subtotal *= 0.9

    total = int(subtotal)
    
    if total < 50000 and km > 5:
        total = 50000
        
    return total
```

```python
def angle_abc(x1, y1, x2, y2, x3, y3):
    a = ((x3-x2)**2 + (y3-y2)**2)**0.5
    c = ((x1-x2)**2 + (y1-y2)**2)**0.5
    b2 = (x1-x3)**2 + (y1-y3)**2
    cos_b = (a**2 + c**2 - b2) / (2*a*c + 1e-9)
    
    low, high = 0.0, 180.0
    for _ in range(50):
        mid = (low + high) / 2
        rad = mid * 3.141592653589793 / 180
        cos_mid = 1 - rad**2/2 + rad**4/24 - rad**6/720 + rad**8/40320
        if cos_mid > cos_b: 
            low = mid
        else: 
            high = mid
    return (low + high) / 2
```

#### Câu 25
Yêu cầu: Viết hàm point_in_quadrilateral(quad, p) trả về True nếu điểm p nằm bên trong hoặc trên biên của tứ giác quad, và False nếu ngược lại. 
Đầu vào: 
- quad: danh sách hoặc bộ gồm đúng 4 đỉnh [(x1,y1), (x2,y2), (x3,y3), (x4,y4)] theo thứ tự (tứ giác có thể lồi hoặc lõm, nhưng không tự cắt). 
- p: một điểm (x, y). - Đầu ra: giá trị boolean. 
- Gợi ý: Cài đặt phép kiểm tra điểm trong đa giác tổng quát (ví dụ: ray casting), đồng thời tính cả trường hợp điểm nằm 
trên biên là ở trong.

```python
def point_in_quadrilateral(quad, p):
    def cross(o, a, b):
        return (a[0] - o[0]) * (b[1] - o[1]) - (a[1] - o[1]) * (b[0] - o[0])
    inside = True
    sign = None
    for i in range(4):
        a = quad[i]
        b = quad[(i + 1) % 4]
        cp = cross(a, b, p)
        if cp != 0: 
            if sign is None:
                sign = cp > 0
            elif (cp > 0) != sign:
                inside = False
                break
    return inside
```

#### Câu 30
Yêu cầu: Viết hàm detect_number_type(s) trả về 'int' nếu s biểu diễn một số nguyên, 'float' nếu biểu diễn một số thực dấu 
chấm động, ngược lại trả về 'neither'. 

Đầu vào: chuỗi s. - Đầu ra: một trong các giá trị {'int', 'float', 'neither'}. 

Quy tắc: 
- Cho phép có dấu + hoặc - ở đầu. 
- Số nguyên phải chỉ gồm các chữ số và không có dấu chấm thập phân (ví dụ: '100'). 
- Số thực phải có dấu chấm và có chữ số ở cả hai phía của dấu chấm (ví dụ: '100.0', '-3.14'). 
- Các chuỗi như '.5', '5.', '1e3' được xem là 'neither' trong bài này. 

```python
def detect_number_type(s):
    if not s: 
        return 'neither'
    s_clean = s[1:] if s[0] in '+-' else s
    if s_clean.isdigit(): 
        return 'int'
    if '.' in s_clean:
        l, r = s_clean.split('.', 1)
        if l.isdigit() and r.isdigit(): return 'float'
    return 'neither'
```

#### Câu 33
Yêu cầu: Viết hàm compute(a, b, op) tính a (+|-||/) b tùy theo op. 
- Đầu vào: các số a, b và ký tự op thuộc {'+', '-', '', '/'}. 
- Đầu ra: kết quả số học. 
- Yêu cầu thêm: - Nếu op là '/' và b == 0, phát sinh ValueError. 
- Nếu op không hợp lệ, phát sinh ValueError. 

```python
def compute(a, b, op):
    if op == '+': return a + b
    if op == '-': return a - b
    if op == '*': return a * b
    if op == '/':
        if b == 0: raise ValueError
        return a / b
    raise ValueError
```

```python
def to_lower_case(c):
    if len(c) != 1: raise ValueError
    return chr(ord(c) + 32) if 'A' <= c <= 'Z' else c
```

#### Câu 41
Viết hàm trả về ký tự đứng sau ký tự x n vị trí trong bảng chữ cái. 
Giữ nguyên chữ hoa/chữ thường và quay vòng trong phạm vi A..Z hoặc a..z. 

```python
def shift_char(x, n):
    if 'a' <= x <= 'z': return chr((ord(x) - 97 + n) % 26 + 97)
    if 'A' <= x <= 'Z': return chr((ord(x) - 65 + n) % 26 + 65)
    return x
```

#### Câu 43
Viết hàm tính diện tích tam giác có ba cạnh a, b, c (dùng công thức Heron). Phát sinh ValueError nếu ba cạnh không thể 
tạo thành tam giác. 

```python
def heron_area(a, b, c):
    if not is_triangle(a, b, c): raise ValueError
    p = (a + b + c) / 2
    return (p * (p - a) * (p - b) * (p - c)) ** 0.5
```

#### Câu 44
Viết hàm trả về tích các số nguyên từ 1 đến n (giai thừa). Với n=0, trả về 1.

```python
def factorial(n):
    res = 1
    for i in range(1, n + 1): res *= i
    return res
```

#### Câu 45
Viết hàm trả về số lượng số nguyên dương k sao cho 1 <= k < n và k > (n - 1) / 2. (Ví dụ: n=9 -> 4).

```python
def count_k(n):
    return sum(1 for k in range(1, n) if k > (n - 1) / 2)
```
