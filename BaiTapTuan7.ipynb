```python
def binary_search(arr, left, right, target):    
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] < target:
            left = mid + 1
        if arr[mid] > target:
            right = mid
        if arr[mid] == target:
            return mid
    return -1
arr = list(map(int, input().split()))
target = int(input())
left = 0
right = len(arr) - 1
print(binary_search(arr, 0, len(arr) - 1, target))
         
```

#### W7A2

số nguyên x, đếm xem x 
xuất hiện bao nhiêu lần 
trong danh sách. def 
count_occurrences(arr, x) -> int: 

```python
def count_occurrences(arr, x): 
    a = 0
    for i in arr:
        if i == x:
            a += 1
    return a
arr = list(map(int, input().split()))
x = int(input())
print(count_occurrences(arr, x))
```

#### W7A3
Viết hàm thực hiện thuật 
toán bubble sort, Selection 
Sort để sắp xếp danh sách 
theo thứ tự tăng dần. Sắp 
xếp nổi bọt (Bubble Sort); Sắp xếp chọn (Selection 
Sort) 

```python
def Bubble_Sort(a):
    n = len(a)
    for i in range(n):
        for j in range(0, n - i - 1):
            if a[j] > a[j + 1]:
                a[j], a[j + 1] = a[j + 1], a[j]
    return a

def Selection_Sort(a):
    n = len(a)
    for i in range(n):
        m = i
        for j in range(i + 1, n):
            if a[j] < a[m]:
                m = j
        a[i], a[m] = a[m], a[i]
    return a

a = list(map(int, input().split()))
print(Bubble_Sort(a))
print(Selection_Sort(a))
```

#### W7A4
Viết hàm tìm phần tử có số 
lần xuất hiện nhiều nhất 
trong danh sách. Nếu có 
nhiều phần tử cùng số lần, 
trả về phần tử xuất hiện 
trước. 

```python
def main(a):
    dic = {}
    k = 0
    for i in a:
        if i not in dic:
            dic[i] = 1
        else:
            dic[i] += 1
        if dic[i] > k:
            k = dic[i]
    for i in dic:
        if dic[i] == k:
            return i
a = list(map(int, input().split()))
print(main(a))
        
```

#### W7A5
Cho danh sách các số 
nguyên và một số X, đếm 
tất cả các cặp (a, b) sao 
cho a + b = X. c 

```python
def main(l, x):
    d = {}
    c = 0
    for i in l:
        t = x - i
        if t in d:
            c += d[t]
        d[i] = d.get(i, 0) + 1
    return c
A = list(map(int, input().split()))
X = int(input())
print(main(A, X))

        
```

phẳng danh sách.

#### W7A7
Cho một danh sách các số 
nguyên nums, hãy tìm độ 
dài của dãy con tăng 
nghiêm ngặt dài nhất 
(không nhất thiết phải liền 
kề). 

```python
def main(a):
    r = []
    for v in a:
        i, j = 0, len(r)
        while i < j:
            m = (i + j) // 2
            if r[m] < v:
                i = m + 1
            else:
                j = m
        if i == len(r):
            r.append(v)
        else:
            r[i] = v
    return len(r)
A = list(map(int, input().split()))
print(main(A))
```

#### W7A8
Bạn được cung cấp: một 
danh sách các khoảng thời 
gian interval = [[l1, r1], [l2, 
r2], ...] và một danh sách 
các truy vấn queries = [q1, 
q2, ...]  
Với mỗi truy vấn q, hãy tìm 
độ dài nhỏ nhất của 
khoảng thời gian [li, ri] sao 
cho li ≤ q ≤ ri.  
Nếu không tồn tại khoảng 
thời gian nào như vậy, hãy 
trả về -1 cho truy vấn đó. 
solve in < O(m.n), hint: 
binary search in the 
queries list 

```python
def main(inv, qrs):
    m = len(qrs)
    q = sorted([(v, i) for i, v in enumerate(qrs)])
    v = sorted(inv, key=lambda x: x[1] - x[0] + 1)
    r = [-1] * m
    p = list(range(m + 1))

    def g(i):
        while p[i] != i:
            p[i] = p[p[i]]
            i = p[i]
        return i

    for a, b in v:
        x, y = 0, m
        while x < y:
            mid = (x + y) // 2
            if q[mid][0] < a:
                x = mid + 1
            else:
                y = mid
        
        j = g(x)
        while j < m and q[j][0] <= b:
            r[q[j][1]] = b - a + 1
            p[j] = j + 1
            j = g(j)

    return r
inv = [tuple(map(int, input().split())) for _ in range(int(input()))]
qrs = list(map(int, input().split()))
print(main(inv, qrs))
    
```

#### W7A9
Cho một danh sách các số 
nums, với mỗi phần tử 
nums[i], hãy đếm xem có 
bao nhiêu phần tử bên phải 
nhỏ hơn nums[i].  
Trả về danh sách các số 
đếm. 

```python
def main(a):
    n = len(a)
    r = [0] * n
    
    def s(e):
        if len(e) <= 1:
            return e
        
        m = len(e) // 2
        x = s(e[:m])
        y = s(e[m:])
        
        i, j = 0, 0
        t = []
        
        while i < len(x) and j < len(y):
            if x[i][1] <= y[j][1]:
                r[x[i][0]] += j
                t.append(x[i])
                i += 1
            else:
                t.append(y[j])
                j += 1
                
        while i < len(x):
            r[x[i][0]] += j
            t.append(x[i])
            i += 1
            
        t.extend(y[j:])
        return t

    s(list(enumerate(a)))
    return r
A = list(map(int, input().split()))
print(main(A))
```

#### W7A10
Cho một danh sách các số 
nguyên nums và một số 
nguyên m, hãy tìm tổng 
mảng con lớn nhất theo 
modulo m.  
Tức là, với bất kỳ mảng 
con nums[i:j] nào, hãy tính 
sum(nums[i:j]) % m và trả 
về kết quả lớn nhất có thể.

```python
def main(a, m):
    r, p = 0, 0
    s = [0]
    for v in a:
        p = (p + v) % m
        i, j = 0, len(s)
        while i < j:
            k = (i + j) // 2
            if s[k] <= p:
                i = k + 1
            else:
                j = k
        if i < len(s):
            r = max(r, (p - s[i] + m) % m)
        r = max(r, p)
        
        i, j = 0, len(s)
        while i < j:
            k = (i + j) // 2
            if s[k] < p:
                i = k + 1
            else:
                j = k
        s.insert(i, p)
    return r
A = list(map(int, input().split()))
M = int(input())
print(main(A, M))
```

#### W7A13
Dãy hài hòa là một dãy 
thỏa mãn điều kiện giá trị 
lớn nhất và giá trị nhỏ nhất 
của dãy khác nhau đúng 
bằng \( 1 \). 
Viết hàm trả về độ dài của 
dãy hài hòa con dài nhất có 
thể có trong mảng giá trị 
nguyên \( arr \) có \( n \) 
phần tử 
Ví dụ: 
Input: 1 3 2 2 5 2 3 7 
Output: 5 - Dãy hài hòa con 
dài nhất là [3,2,2,2,3] 

```python
def main(a):
    d = {}
    for x in a:
        d[x] = d.get(x, 0) + 1
    
    m = 0
    for k in d:
        if k + 1 in d:
            c = d[k] + d[k + 1]
            if c > m:
                m = c
                
    return m
A = list(map(int, input().split()))
print(main(A))
```
