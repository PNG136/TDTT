#### W7A16
Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu 
mỗi phần tử x trong mảng xuất hiện đúng x lần. 
Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa 
đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp. 

```python
def main(A):
    dic = {}
    for i in A:
        if i in dic:
            dic[i] += 1
        else: 
            dic[i] = 1
    for i in dic:
        if int(i) != dic[i]:
            k = int(i) - dic[i]
            if k > 0:
                for n in range(k):
                    A.append(int(i))
            if k < 0:
                for n in range(-k):
                    A.remove(int(i))
    return A

A = list(map(int, input().split()))
print(main(A))
```

#### W7A17
Tìm phần tử xuất hiện nhiều hơn n/2 lần trong mảng (nếu có) với độ 
phức tạp thời gian là O(n), độ phức tạp bộ nhớ là O(1), - Boyer–Moore. 
Nếu không có phần tử nào thỏa mãn thì không in gì 

```python
def main(A):
    n = len(A)
    a = 0
    b = None
    if n == 0:
        return
    for i in A:
        if a == 0:
            b = i
            a = 1
        elif b == i:
            a += 1
        else:
            a -= 1
    if a <= 0:
        return
    else:
        a = 0
        for i in A:
            if i == b:
                a += 1
        if a > n // 2:
            return b
A = list(map(int, input().split()))
print(main(A))
```

#### W7A19
Merge 2 mảng đã sắp xếp (có thể tăng dần, có thể giảm dần) mà không 
dùng bất cứ thuật toán sắp xếp nào 

```python
def main(A, B):
    n = len(A)
    m = len(B)
    C = []

    if n == 0:
        return B
    if m == 0:
        return A
    if n == 1 and m == 1:
        return A + B
    
    i = 0 if A[0] <= A[-1]  else n - 1
    r = 1 if A[0] <= A[-1]  else -1

    j = 0 if B[0] <= B[-1]  else m - 1
    q = 1 if B[0] <= B[-1]  else -1

    while 0 <= i < n and 0 <= j < m:
        if A[i] <= B[j]:
            C.append(A[i])
            i += r
        else:
            C.append(B[j])
            j += q
    if n < m:
        while 0 <= j < m:
            C.append(B[j])
            j += q
    if n > m:
        while 0 <= i < n:
            C.append(A[i])
            i += r
    return C
    
A = list(map(int, input().split()))
B = list(map(int, input().split()))
print(main(A, B))

```

#### W7A27
 
Cho một dãy ngoặc ()(){}{}[][]..., kiểm tra xem dãy ngoặc đã cho có hợp lệ 
hay không 

```python
def main(n):
    dic = {'(': ')', '[': ']', '{': '}'}
    if len(n) % 2 != 0:
        return False
    for i in range(0, len(n), 2):
        if n[i] in dic:
            if n[i + 1] != dic[n[i]]:
                return False
    return True
n = input()
print(main(n))

    
            
```

#### W7A28
Bài toán Tìm dãy các hợp số dài nhất liên tiếp không chứa số nguyên 
nào <= N (N là số nguyên dương cho trước). Nếu có nhiều dãy cùng độ 
dài thì lấy dãy nhỏ nhất (gặp đầu tiên từ trái qua) 

```python
def main(n):
    if n < 4:
        return []
    A = [True] * (n + 1)
    A[0] = A[1] = False
    for i in range(2, int(n ** 0.5) + 1):
        if A[i]:
            for j in range(i * i, n + 1, i):
                A[j] = False
    a = 3
    b = 3
    s = 0
    k = 0
    for i in range(4, n + 1):
        if not A[i]:
            if s == 0:
                a = i
            s += 1
        else:
            if k < s:
                k = s
                b = a
            s = 0              
    return list(range(b , b + k))

n = int(input())
print(main(n))
```

#### W7A29
Tìm Dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ 
bàn phím. Lấy dãy đầu tiên nếu có nhiều dãy có số phần tử bằng nhau

```python
def main(A):
    n = len(A)
    if n <= 2:
        return sorted(A)
    a = 0
    c = 0
    d = 0
    
    for i in range(1, n):
        if A[i] >= A[i - 1]:
            d += 1
        else:
            if c < d:
                c = d
                a = i - d - 1
            d = 0
    return A[a:a + c + 1]

A = list(map(int, input().split()))
print(main(A))
```

#### W7A30
Tìm Dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập 
vào từ bàn phím

```python
def main(a):
    r = []
    for v in a:
        i, j = 0, len(r)
        while i < j:
            m = (i + j) // 2
            if r[m] < v:
                i = m + 1
            else:
                j = m
        if i == len(r):
            r.append(v)
        else:
            r[i] = v
    return len(r)
A = list(map(int, input().split()))
print(main(A))
```
