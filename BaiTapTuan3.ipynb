[Phân loại tam giác] 
Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm tra 3 số đã nhập có tạo thành một tam giác hợp 
lệ hay không. Nếu không in ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay 
"Tam giác thường". 

```python
a = float(input())
b = float(input())
c = float(input())
if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print('tam giác đều')
    elif a == b or a == c or b == c:
        print('tam giác cân')
    else:
        if a**2 == b**2 + c**2 or b**2 == a**2 + c**2 or c**2 == a**2 + b**2:
            print('tam giác vuông')
        else:
            print('tam giác thường')
    p = (a + b + c) / 2
    print(f'diện tích tam giác là: {p * (p - a) * (p - b) * (p - c)}')
    print(f'chu vi tam giác là: {a + b + c}')
else:
    print('không là 3 cạnh của một tam giác')
```

#### W3A14
[Giải phương trình bậc nhất] 
Nhập vào hai số thực a và b của phương trình ax + b = 0. Tìm và in ra nghiệm của phương trình, kết quả làm tròn đến số 
thập phân thứ 2. In ra "Vô nghiệm" hoặc "Vô số nghiệm" trong trường hợp tương ứng. 

```python
a, b, c = map(float, input().split())
delta = b**2 - 4*a*c
if delta < 0:
    print('phương trình vô nghiệm')
else :
    x1 = (-b + delta**0.5) / (2*a)
    x2 = (-b - delta**0.5) / (2*a)
    print(f'x1 = {x1}, x2 = {x2}')
```
