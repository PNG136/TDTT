#### W6A16
 
Nhập vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma 
trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải 

```python
A = []
n = int(input())
for i in range(n):
    A.append(list(map(int, input().split())))
for row in A:
    for val in row:
        print(f"{val}", end="    ") 
    print()
```

#### W6A17
Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và 
đường chéo phụ của mảng 

```python
n = int(input())
A = [list(map(int, input().split())) for _ in range(n)]
print("Đường chéo chính:", *(A[i][i] for i in range(n)))
print("Đường chéo phụ:", *(A[i][n - 1 - i] for i in range(n)))
```

#### W6A18
Nhập vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên 
cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột 
k (hàng k) 

```python
n, m, k = map(int, input().split())
a = [list(map(int, input().split())) for _ in range(n)]

sum_row = sum(a[k]) if k < n else 0
sum_col = sum(a[i][k] for i in range(n)) if k < m else 0

print(f"Tổng hàng {k}: {sum_row}")
print(f"Tổng cột {k}: {sum_col}")
```

#### W6A20
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và 
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về -1

```python
A = list(map(int, input().split()))
k = int(input())
for i in range(len(A)):
    if A[i] == k:
        print(i)
        break
else:
    print(-1)
```

#### W6A21
Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ 
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện 

```python
A = input().split()
dic = {}
for i in range(len(A)):
    if A[i] not in dic:
        dic[A[i]] = () + (i,)
    else:
        dic[A[i]] += (i,)
print(dic)
```

#### W6A23
VIết hàm xóa đi các số < x có trong mảng 

```python
A = list(map(int, input().split()))
x = int(input())
a = 0
while a != len(A):
    if A[a] < x:
        A.remove(A[a])
        continue
    a += 1
print(A)
```

#### W6A24
Viết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí 
trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 0 2 3 2 4 9 

```python
A = list(map(int, input().split()))
k = 0
a = 0
while a != len(A):
    if A[a] == 0:
        k += 1
        A.remove(0)
        continue
    a += 1
print([0] * k + A)        
```

#### W6A26
Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2 
phần tử đầu và cuối 

```python
A = list(map(int, input().split()))
k = 0
for i in range(1, len(A) - 1):
    if A[i-1] < A[i] and A[i+1] < A[i]:
        k += A[i]
print(k)
```

#### W6A28
Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm 
thực sự, hay vô danh (không tăng, không giảm) 

```python
A = list(map(int, input().split()))
k = 0
for i in range(1, len(A)):
    if A[i-1] < A[i]:
        k += 1
    elif A[i-1] > A[i]:
        k -= 1
    else:
        continue
if k == len(A)-1:
    print('dãy tăng')
elif k == -(len(A)-1):
    print('dãy giảm')
else:
    print('vô danh')
```
