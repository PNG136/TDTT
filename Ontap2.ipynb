#### Bài 6
Yêu cầu: Cho số nguyên không âm n, hãy tạo số lớn nhất và số nhỏ nhất có thể tạo được bằng cách sắp xếp lại các chữ số của n.  
Trả về dưới dạng danh sách [max_value, min_value] (không dùng tuple).  
Với số nhỏ nhất, không được để chữ số 0 đứng đầu (ví dụ: 2015 → 5210 và 1025). 

```python
def max_min_digits(n):
    d = sorted(str(n))
    mx = int("".join(d[::-1]))
    if d[0] == '0' and len(d) > 1:
        for i in range(len(d)):
            if d[i] != '0':
                d[0], d[i] = d[i], d[0]
                break
    mn = int("".join(d))
    return [mx, mn]
```

#### Bài 8
Yêu cầu: Cho chuỗi thời gian dạng "hh:mm:ss", hãy trả về tổng số giây.  
Giờ, phút, giây phải là số nguyên với 0 ≤ mm < 60 và 0 ≤ ss < 60.  
Trả về -1 nếu giá trị nằm ngoài phạm vi hợp lệ. 

```python
def time_to_seconds(s):
    try:
        h, m, sec = map(int, s.split(':'))
        if 0 <= m < 60 and 0 <= sec < 60:
            return h * 3600 + m * 60 + sec
        return -1
    except:
        return -1
```

#### Bài 26
Viết hàm `sum_divisors(a)` trả về tổng của tất cả các ước dương của a. Nếu $a \le 0$ thì trả về 0.

```python
def sum_divisors(a):
    if a <= 0: return 0
    s = 0
    for i in range(1, int(a**0.5) + 1):
        if a % i == 0:
            s += i
            if i != a // i: s += a // i
    return s
```

#### Bài 27
Viết hàm `is_perfect(n)` trả về True nếu n là số hoàn hảo và False nếu không phải. Một số nguyên dương được gọi là hoàn hảo nếu nó bằng tổng các ước dương thực sự của nó (không tính chính nó). Theo định nghĩa, n phải dương; với giá trị không dương thì trả về False.

```python
def is_perfect(n):
    if n <= 1: return False
    s = 1
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            s += i + (n // i if i != n // i else 0)
    return s == n
```

#### Bài 30
Yêu cầu: Cài đặt giải mã Atbash cho các chữ cái tiếng Anh[cite: 4]. Ánh xạ a $\rightarrow$ z, b $\rightarrow$ y, ..., z $\rightarrow$ a và tương tự A $\rightarrow$ Z, ..., Z $\rightarrow$ A[cite: 4]. Các ký tự không phải chữ cái được giữ nguyên[cite: 4].

```python
def atbash_cipher(s):
    r = []
    for c in s:
        if 'a' <= c <= 'z': r.append(chr(219 - ord(c))) # 97+122 = 219
        elif 'A' <= c <= 'Z': r.append(chr(155 - ord(c))) # 65+90 = 155
        else: r.append(c)
    return "".join(r)
```

#### Bài 31
Yêu cầu: Một chuỗi được gọi là perfect nếu nó chứa ít nhất một chữ hoa, ít nhất một chữ thường và ít nhất một chữ số. Trả về True nếu chuỗi là perfect; ngược lại trả về False. Nếu đầu vào không phải chuỗi, hãy chuyển thành chuỗi.

```python
def is_perfect_string(s):
    s = str(s)
    u = l = d = False
    for c in s:
        if c.isupper(): u = True
        elif c.islower(): l = True
        elif c.isdigit(): d = True
    return u and l and d
```

#### Bài 32
Liệt kê tất cả số nguyên tố nhỏ hơn x (x < 1000). Viết hàm `primes_less_than(x)` trả về danh sách tất cả các số nguyên tố nhỏ hơn x. 
- Dùng các vòng lặp đơn giản và phép kiểm tra số nguyên.
- Không dùng tuple hoặc dictionary.

```python
def primes_less_than(x):
    r = []
    for n in range(2, x):
        p = True
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                p = False
                break
        if p: r.append(n)
    return r
```

Nếu không thể, trả về (False, []). Nếu có thể, trả về (True, a).

```python
def build_even_odd_array(n):
    k = n // 2
    if k % 2 != 0: return (False, [])
    a = [2 * i for i in range(1, k + 1)]
    b = [2 * i - 1 for i in range(1, k)]
    b.append(sum(a) - sum(b))
    return (True, a + b)
```

```python
def inverted_triangle(n):
    if n <= 0 or n % 2 == 0: return ""
    return "\n".join(" " * i + "*" * (n - 2 * i) for i in range(n // 2 + 1))
```

#### Bài 39
Tổng của hai đường chéo trong ma trận n x n. Viết hàm `sum_two_diagonals(mat)` trả về tổng các phần tử trên đường chéo chính và đường chéo phụ. Nếu n là số lẻ thì không được cộng trùng ô ở giữa. 
- Dùng vòng lặp. 
- Không dùng tuple hoặc dictionary.

```python
def sum_two_diagonals(mat):
    n = len(mat)
    s = 0
    for i in range(n): s += mat[i][i] + mat[i][n - 1 - i]
    if n % 2 != 0: s -= mat[n // 2][n // 2]
    return s
```

```python
def sum_where_i_plus_j_even(mat):
    s = 0
    for i in range(len(mat)):
        for j in range(len(mat[0])):
            if (i + j) % 2 == 0: s += mat[i][j]
    return s
```

```python
def sum_odd_rows_or_cols(mat):
    s = 0
    for i in range(len(mat)):
        for j in range(len(mat[0])):
            if i % 2 == 0 or j % 2 == 0: s += mat[i][j]
    return s
```

#### Bài 42
Trả về cột (chỉ số bắt đầu từ 0) có tổng cột lớn nhất trong ma trận n x m. Viết hàm `column_with_max_sum(mat)` trả về chỉ số bắt đầu từ 0 của cột có tổng lớn nhất. Nếu có nhiều cột cùng tổng lớn nhất, trả về cột có chỉ số nhỏ nhất. 
- Dùng vòng lặp để tính tổng từng cột.
- Không dùng tuple hoặc dictionary.

```python
def column_with_max_sum(mat):
    r, c = len(mat), len(mat[0])
    mx, idx = float('-inf'), -1
    for j in range(c):
        s = sum(mat[i][j] for i in range(r))
        if s > mx:
            mx, idx = s, j
    return idx
```

#### Bài 43
Viết hàm `zigzag_numbers(n)` trả về một chuỗi biểu diễn ma trận n x n chứa các số từ 1 đến $n^2$. Các hàng cách nhau bởi ký tự xuống dòng. Với chỉ số hàng bắt đầu từ 0: hàng chẵn đi từ trái sang phải, hàng lẻ đi từ phải sang trái. Các số trong cùng một hàng cách nhau bởi một dấu cách.

```python
def zigzag_numbers(n):
    res = []
    for i in range(n):
        row = list(range(i * n + 1, (i + 1) * n + 1))
        if i % 2 != 0: row.reverse()
        res.append(" ".join(map(str, row)))
    return "\n".join(res)
```

#### Bài 44
Viết hàm `count_solutions_3vars(a, b, c, t)` trả về số nghiệm nguyên không âm (x, y, z) của phương trình $a*x + b*y + c*z = t$. [cite_start]Giả sử a, b, c, t là các số nguyên. [cite_start]Nếu $a \le 0$ hoặc $b \le 0$ hoặc $c \le 0$ hoặc $t < 0$ thì trả về 0.

```python
def count_solutions_3vars(a, b, c, t):
    if a <= 0 or b <= 0 or c <= 0 or t < 0: return 0
    cnt = 0
    for x in range(t // a + 1):
        for y in range((t - a * x) // b + 1):
            if (t - a * x - b * y) % c == 0: cnt += 1
    return cnt
```

#### Bài 45
Số lần đổi chỗ kề nhau ít nhất để đưa số 1 duy nhất vào giữa ma trận có kích thước lẻ. Bạn được cho ma trận n x n (với n là số lẻ). Ma trận chứa đúng một ô có giá trị 1, còn tất cả các ô khác đều là 0[cite: 6]. Trong một thao tác, bạn có thể đổi chỗ hai hàng kề nhau hoặc hai cột kề nhau. Hãy tính số thao tác ít nhất để đưa số 1 vào đúng ô trung tâm.

```python
def min_swaps_to_center(mat):
    n = len(mat)
    for i in range(n):
        for j in range(n):
            if mat[i][j] == 1:
                return abs(i - n // 2) + abs(j - n // 2)
    return 0
```

#### Bài 46
Yêu cầu: Cho danh sách tên sinh viên, hãy trả về 3 tên xuất hiện nhiều nhất. Nếu có ít hơn 3 tên khác nhau, hãy trả về tất cả. 
- Yêu cầu thêm: Ưu tiên dùng map/dictionary.
- Khai báo hàm: `top3_frequent_names(names) -> list[str]`.

```python
def top3_frequent_names(names):
    d = {}
    for n in names: d[n] = d.get(n, 0) + 1
    return sorted(d.keys(), key=lambda k: (-d[k], k))[:3]
```

#### Bài 47
Yêu cầu: Cho danh sách các bộ ba (tuple gồm 3 số nguyên), hãy trả về bộ ba xuất hiện nhiều nhất. 
- Khai báo hàm: `most_frequent_triple(triples) -> tuple[int,int,int]`. 
- Phát sinh ValueError nếu danh sách đầu vào rỗng.

```python
def most_frequent_triple(triples):
    if not triples: raise ValueError
    d = {}
    for t in triples: d[t] = d.get(t, 0) + 1
    return max(d, key=d.get)
```

#### Bài 48
Cho danh sách số nguyên nums và số nguyên k, hãy trả về True nếu tồn tại hai chỉ số khác nhau i và j sao cho: 
- `nums[i] == nums[j]`, và
- `abs(i - j) <= k`.
Khai báo hàm: `contains_nearby_duplicate(nums, k) -> bool`.

```python
def contains_nearby_duplicate(nums, k):
    d = {}
    for i, v in enumerate(nums):
        if v in d and i - d[v] <= k: return True
        d[v] = i
    return False
```

#### Bài 49
Bạn được cho hai mảng số nguyên 2 chiều items1 và items2. Mỗi phần tử có dạng `[value, weight]`, và value là duy nhất trong từng danh sách. Yêu cầu: Trả về danh sách 2 chiều ret, trong đó mỗi phần tử là `[value, total_weight]`, với total_weight là tổng trọng lượng của cùng value trong cả hai danh sách.  quả cần được sắp xếp theo value tăng dần.

```python
def merge_items(items1, items2):
    d = {}
    for v, w in items1 + items2:
        d[v] = d.get(v, 0) + w
    return sorted([[v, w] for v, w in d.items()])
```
