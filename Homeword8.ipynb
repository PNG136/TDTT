#### W8A16
Yêu cầu 
Xây dựng lớp Shape với phương thức area(). 
Xây dựng lớp Rectangle(w, h) và Circle(r) kế thừa từ Shape, ghi đè area(). 
Lấy π = 3.14. 
 
Input 
Dòng 1: loại hình (RECTANGLE hoặc CIRCLE) 
Dòng 2: tham số tương ứng. 
 
Output 
In ra diện tích với 2 chữ số thập phân. 

```python
class Shape:
    def area(seft):
        return 0.0

class Rectangle(Shape):
    def __init__(seft, w, h):
        seft.w = w
        seft.h = h
    
    def area(seft):
        return seft.w * seft.h

class Circle(Shape):
    def __init__(seft, r):
        seft.r = r

    def area(seft):
        return 3.14 * seft.r * seft.r

a = input().strip().upper()

if a == 'RECTANGLE':
    w, h = map(float, input().split())
    shape = Rectangle(w, h)
    print(f"{shape.area():.2f}")

elif a == 'CIRCLE':
    r = float(input())
    shape = Circle(r)
    print(f"{shape.area():.2f}")
```

Xây dựng lớp Polygon với phương thức perimeter(). 
Xây dựng lớp Triangle(a, b, c) kế thừa từ Polygon và ghi đè perimeter() trả về a+b+c. 
 
Input 
Một dòng chứa 3 số nguyên a b c. 
 
Output 
In ra chu vi tam giác. 

```python
class Polygon:
    def perimeter():
        return 0.0
    
class Triangle(Polygon):
    def __init__(seft, a, b, c):
        seft.a = a
        seft.b = b
        seft.c = c

    def perimeter(seft):
        return seft.a + seft.b + seft.c
    
a, b, c = map(float,input().split())

Triangle = Triangle(a, b, c)
print(Triangle.perimeter())
```

Xây dựng lớp SavingAccount(balance, rate) kế thừa từ Account, thêm phương thức 
add_interest() để cộng balance*rate vào số dư. 
 
Input 
balance rate 
 
Output 
Số dư sau khi cộng lãi, in 2 chữ số thập phân. 

```python
class Account:
    def __init__(self, balance):
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount

    def get_balance(self):
        return self.balance

class SavingAccount(Account):
    def __init__(self, balance, rate):
        super().__init__(balance)
        self.rate = rate

    def add_interest(self):
        self.balance = self.balance * self.rate + self.balance

balance, rate = map(float, input().strip().split())

saving_account = SavingAccount(balance, rate)

saving_account.add_interest()

print(saving_account.get_balance())


        
```

Xây dựng lớp Printer với phương thức print_value(x). 
Xây dựng lớp BinaryPrinter kế thừa từ Printer, ghi đè print_value(x) để in biểu diễn nhị phân của 
số nguyên dương x. 
 
Input 
Một số nguyên dương n. 
 
Output 
In ra dạng nhị phân của n.

```python
class Printer:
    def print_value(self):
        pass

class BinaryPrinter(Printer):
    def __init__(self, x):
        self.x = x

    def print_value(self):
        k = bin(x)
        return k[2:len(k)]
    
x = int(input())

binary = BinaryPrinter(x)

print(binary.print_value())
```

Xây dựng lớp Matrix2x2(a, b, c, d) biểu diễn ma trận: 
[a b] 
[c d] 
Viết phương thức trace() trả về a + d và determinant() trả về ad - bc. 
 
Input 
Một dòng chứa 4 số nguyên a b c d. 
 
Output 
In ra trace và determinant trên cùng một dòng. 

```python
class Matrix2x2:
    def __init__(self, a, b, c, d):
        self.a = a
        self.b = b
        self.c = c
        self.d = d

    def trace(self):
        return self.a + self.d
    
    def deterninant(self):
        return self.a * self.d - self.b * self.c
    
a, b, c, d = map(int, input().split())

matrix = Matrix2x2(a, b, c, d)

print(matrix.trace(), matrix.deterninant())
```

Yêu cầu 
Xây dựng lớp Product(name, price). 
Xây dựng lớp DiscountProduct(name, price, percent) kế thừa từ Product, ghi đè phương thức 
final_price() để trả về giá sau giảm. 
Công thức: price * (100 - percent) / 100 
 
Input 
name price percent 
Output 
In ra giá cuối cùng với 2 chữ số thập phân. 

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

class DiscountProduct(Product):
    def __init__(self, name, price, percent):
        super().__init__(name, price)
        self.percent = percent

    def final_price(self):
        return float(self.price) * (100 - float(self.percent)) / 100
    
a, b, c = input().split()

discount = DiscountProduct(a, b, c)

print(f'{discount.final_price():.2f}')
```

#### W8A28
Yêu cầu 
Xây dựng lớp Counter có thuộc tính count ban đầu bằng 0 và các phương thức inc(), dec(), 
value(). 
Sau đó thực hiện dãy thao tác trên đối tượng. 
 
Input 
Dòng 1: n 
Dòng 2: n thao tác, mỗi thao tác là INC hoặc DEC, cách nhau bởi dấu cách. 
 
Output 
Giá trị cuối cùng của count. 

```python
class Counter:
    def __init__(self):
        self.count = 0

    def inc(self):
        self.count += 1

    def dec(self):
        self.count -= 1

    def value(self):
        return self.count


n = int(input().strip())

operations = input().strip().split()

counter = Counter()

for op in operations:
    if op == "INC":
        counter.inc()
    elif op == "DEC":
        counter.dec()

print(counter.value())
```

#### W8A29

Xây dựng lớp Time(h, m, s) với phương thức to_seconds() đổi thời gian sang tổng số giây. 
 
Input 
h m s 
 
Output 
Tổng số giây. 

```python
class Time:
    def __init__(self, h, m, s):
        self.h = h
        self.m = m
        self.s = s

    def to_seconds(self):
        return self.h * 3600 + self.m * 60 + self.s
    
h, m, s = map(int, input().split())

time = Time(h, m, s)

print(time.to_seconds())
```

#### W8A30
Yêu cầu 
Xây dựng lớp Stack sử dụng danh sách để cài đặt ngăn xếp với các phương thức push(x), 
pop(), top(). 
Dữ liệu đảm bảo mọi lệnh pop và top đều hợp lệ. 
 
Input 
Dòng 1: n 
n dòng tiếp theo là một trong ba lệnh: 
PUSH x 
POP 
TOP 
 
Output 
Mỗi lệnh TOP in ra giá trị ở đỉnh ngăn xếp trên một dòng.

```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, x):
        self.stack.append(x)

    def pop(self):
        if self.stack:
            self.stack.pop()

    def top(self):
        if self.stack:
            return self.stack[-1]
        return None


n = int(input().strip())

my_stack = Stack()

for _ in range(n):
    line = input().strip().split()
    command = line[0]
    
    if command == "PUSH":
        x = int(line[1])  
        my_stack.push(x)
        
    elif command == "POP":
        my_stack.pop()
        
    elif command == "TOP":
        print(my_stack.top())
```
