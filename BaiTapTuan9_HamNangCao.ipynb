#### W9A1
Viết hàm merge hai danh sách đã sắp xếp mà không dùng bất cứ thuật toán sắp xếp nào

```python
def main(A, B):
    n = len(A)
    m = len(B)
    C = []

    if n == 0:
        return B
    
    if m == 0:
        return A
    
    if n == 1 and m == 1:
        return A + B
    
    i = 0 if A[0] <= A[-1]  else n - 1
    r = 1 if A[0] <= A[-1]  else -1

    j = 0 if B[0] <= B[-1]  else m - 1
    q = 1 if B[0] <= B[-1]  else -1

    while 0 <= i < n and 0 <= j < m:
        if A[i] <= B[j]:
            C.append(A[i])
            i += r
        else:
            C.append(B[j])
            j += q

    if n <= m:
        while 0 <= j < m:
            C.append(B[j])
            j += q

    if n >= m:
        while 0 <= i < n:
            C.append(A[i])
            i += r
    return C
    
A = list(map(int, input().split()))
B = list(map(int, input().split()))

print(main(A, B))
```

#### W9A2
Cho ma trận hai chiều nxm có các phần tử được sắp xếp trên dòng tăng dần và trên cột 
tăng dần, hãy tìm số lớn thứ k (k < nxm) với thuật toán tốt nhất có thể 

```python
def count(M, x):

    cnt, c = 0, len(M[0]) - 1
    for r in range(len(M)):
        while c >= 0 and M[r][c] > x:
            c -= 1
        cnt += (c + 1)
    return cnt

def main(M, k):
    n, m = len(M), len(M[0])
    k_small = n * m - k + 1 
    
    L, R = M[0][0], M[-1][-1]
    ans = L
    
    while L <= R:
        mid = (L + R) // 2
        if count(M, mid) >= k_small:
            ans = mid
            R = mid - 1
        else:
            L = mid + 1
    return ans

n = int(input())
k = int(input())
A = []

for i in range(n):
    A.append(list(float, input().split()))

print(main(A, k))
```

#### W9A3
Cho một số nguyên dương, hãy thực hiện hoán đổi đúng 2 chữ số để được một số 
nguyên nhỏ nhất có thể

```python
def main(s):
    last = {s[i]: i for i in range(len(s))}
    lst = list(s)

    for i, c in enumerate(s):
        for digit in sorted(last.keys()):
            if digit >= c:
                break
            if last[digit] > i:
                lst[i], lst[last[digit]] = lst[last[digit]], lst[i]
                return int("".join(lst))
            
    lst[-2], lst[-1] = lst[-1], lst[-2]
    
    return int("".join(lst))

s = input()
print(main(s))
```

#### W9A4
Cho xâu a gồm các chữ cái thường và in hoa và xâu b chỉ gồm các chữ cái in hoa. Ta có 
thể biến đổi xâu a theo 2 bước như sau: - Chuyển một số chữ cái thường trong xâu a thành chữ cái in hoa (có thể không biến đổi 
chữ cái nào) - Xóa hết tất cả các chữ cái thường trong xâu a. 
 
Cho 2 xâu a,b, xác đính xem có cách biến đổi từ xâu a sang xâu b không. Ghi ra "YES" 
nếu có và "NO" nếu ngược lại. 
 
Đầu vào: 
Dòng đầu tiên chứa 1 số nguyên Q là số lượng câu hỏi. Tương ứng với mỗi câu hỏi, đưa 
ra 2 xâu a,b trên 2 dòng khác nhau.(Q≤10,|a|,|b|≤1000 
) 
 
Đầu ra:  
Với mỗi câu hỏi, ghi ra kết quả tương ứng trên một dòng.

```python
def check(a, b):
    n, m = len(a), len(b)

    dp = [[False] * (m + 1) for _ in range(n + 1)]
    dp[0][0] = True

    for i in range(n):
        for j in range(m + 1):
            if not dp[i][j]:
                continue

            if a[i].islower():
                dp[i + 1][j] = True

            if j < m and a[i].upper() == b[j]:
                dp[i + 1][j + 1] = True

    return "YES" if dp[n][m] else "NO"

Q = int(input())
for _ in range(Q):
    a = input().strip()
    b = input().strip()
    print(check(a, b))
```

#### W9A5
Cho hai mảng số nguyên list1 và list2, viết hàm kiểm tra xem list1 có thuộc list2 hay không 
(có phải là dãy con liên tiếp của list2 hay không)?

```python
def main(A, B):
    n = len(A)
    m = len(B)
    k = 0
    i = 0

    if n > m:
        return False
    
    while k + n < m and i < n:
        if A[i] == B[k]:
            while i < n - 1:
                i += 1
                k += 1
                if A[i] != B[k]:
                    break
            else:
                return True
            
        else:
            k += 1
            i = 0

    return False

A = list(map(int, input().split()))
B = list(map(int, input().split()))

print(main(A, B))
```

[Hoán vị tiếp theo] 
Cho một dãy số gồm n chữ số là một hoán vị của dãy 1,2,...,n. Viết chương trình in ra 
hoán vị ở vị trí tiếp theo (danh sách các hoán vị được sắp xếp theo thứ tự từ điển) của 
hoán vị đã cho. 
 
Đầu vào từ bàn phím gồm 2 dòng: - Dòng đầu tiên chứa số nguyên n - Dòng tiếp theo chứa n số nguyên là một hoán vị của dãy 1,2,...,n (n≤105) 
 
Đầu ra: In ra màn hình hoán vị tiếp theo của hoán vị đã cho. 

```python
def next_p(a):
    n = len(a)
    i = n - 2

    while i >= 0 and a[i] >= a[i + 1]:
        i -= 1

    if i == -1:
        return a[::-1] 

    j = n - 1
    while a[j] <= a[i]:
        j -= 1

    a[i], a[j] = a[j], a[i]

    a[i + 1:] = reversed(a[i + 1:])
    return a

n = int(input())
a = list(map(int, input().split()))

res = next_p(a)
print(*(res))
```

[CanPlaceFlowers] 
Giả sử bạn là nhân viên của một công ty cây xanh, được giao cho chăm sóc một luống 
đất. Luống đất đó được chia thành nhiều ô đất nhỏ có kích cỡ bằng nhau. Trước khi được 
giao cho bạn chăm sóc, trên luống hoa đã được trồng sẵn một số cây xanh tán rộng. Một 
ngày đầu xuân, công ty gửi đến cho bạn một số cây hoa và yêu cầu bạn trồng vào những 
vị trí còn trống trong luống hoa. Đây là những cây hoa hướng sáng và chúng không thể 
sống được dưới bóng râm của cây khác, đặc biệt là cây tán rộng. Vì vậy, bạn không thể 
trồng chúng tại ô bên cạnh những ô đã được trồng cây xanh tán rộng. Và nếu trồng những 
cây hoa bên cạnh nhau, chúng sẽ tranh giành phân bón và nước và sau cùng, cả hai cây 
sẽ cùng chết. 
 
Viết hàm canPlaceFlowers(flowerbed, k) kiểm tra xem có thể trồng hết số hoa mới được 
cấp vào trong luống đất của bạn hay không. 
 
Hàm nhận đầu vào là một danh sách (list) flowerbed, đánh dấu những vị trí trong luống đất 
đã được trồng cây xanh trước đó (giá trị 1 ứng với vị trí cây đã được trồng và 0 ứng với vị 
trí đất trống) và số lượng hoa được phát thêm k (int). Hàm trả về True nếu có thể trồng hết 
toàn bộ k cây hoa vào luống đất và False trong trường hợp ngược lại. 

```python
def canPlaceFlowers(A, k):
    n = len(A)
    for i in range(n):

        if A[i] == 0:

            prev_ok = (i == 0 or A[i - 1] == 0)
            next_ok = (i == n - 1 or A[i + 1] == 0)
            
            if prev_ok and next_ok:
                A[i] = 1  
                k -= 1 
                
                if k <= 0:
                    return True
                    
    return k <= 0

flowerbed = list(map(int, input().split()))
k = int(input())

print(canPlaceFlowers(flowerbed, k))
```

#### W9A8
[Phát hiện đường đi lặp lại của Rô-bốt] 
 
Ban đầu, một con rô-bốt ở vị trí (0, 0) trên một lưới 2D.  
Cho một chuỗi `moves` chứa các thao tác di chuyển của rô-bốt. 
 
Các bước di chuyển được biểu diễn bởi các ký tự:  - 'R' (Phải) - 'L' (Trái) - 'U' (Lên) - 'D' (Xuống) 
 
Viết hàm `def is_path_crossing(moves)`: 
Hàm nhận đầu vào là một chuỗi `moves`. 
Hàm trả về `True` nếu trong quá trình di chuyển, rô-bốt đi vào một ô mà nó đã từng đi qua 
trước đó.  
Hàm trả về `False` nếu rô-bốt không bao giờ đi lặp lại (luôn đi vào ô mới). 
 
Lưu ý: Vị trí ban đầu (0, 0) được tính là vị trí đầu tiên đã đi qua. 

```python
def is_path_crossing(moves):
    x, y = 0, 0

    vis = {(x, y)}
    
    for m in moves:
        if m == 'U': y += 1
        elif m == 'D': y -= 1
        elif m == 'R': x += 1
        elif m == 'L': x -= 1

        if (x, y) in vis:
            return True

        vis.add((x, y))
        
    return False

moves = list(map(int, input().split()))

print(is_path_crossing(moves))
```

Viết hàm `def Median(nums):` 
Hàm nhận đầu vào là một danh sách (list) `nums` được đảm bảo có đúng 5 số nguyên. 
 
Yêu cầu: 
Tìm và trả về số trung vị (số lớn thứ ba) của 5 số này. 
 
Ràng buộc: 
1. Chỉ được sử dụng tối đa 6 phép so sánh (ví dụ: <, >, <=, >=). 
2. KHÔNG ĐƯỢC DÙNG bất kỳ hàm sắp xếp nào (như `sorted()`, `list.sort()`). 
 
Chú ý: Bạn chỉ cần viết hàm `Median` theo yêu cầu.  
Phần đọc input và gọi hàm (hàm main) đã có sẵn. 

```python
def Median(A):
    def s(i, j):
        if A[i] > A[j]: 
            A[i], A[j] = A[j], A[i]

    s(0, 1)

    s(2, 3)

    s(0, 2)

    s(1, 3)

    s(1, 4)
    
    s(1, 2)

    return A[2]

nums = list(map(int, input().split()))

print(Median(nums))
```

#### W9A10
[Tìm bán kính lò sưởi nhỏ nhất] 
 
Mùa đông đang đến! Bạn được cung cấp vị trí của các ngôi nhà và các lò sưởi trên một 
đường thẳng. 
Nhiệm vụ của bạn là tìm bán kính r nhỏ nhất cho các lò sưởi sao cho mọi ngôi nhà đều 
nằm trong phạm vi sưởi ấm của ít nhất một lò sưởi. 
 
Tất cả các lò sưởi đều có cùng một bán kính r. 
Một ngôi nhà tại vị trí h được sưởi ấm bởi một lò sưởi tại vị trí t nếu |h - t| <= r. 
 
Viết hàm `def findRadius(houses, heaters):` 
Hàm nhận đầu vào là hai danh sách (list) các số nguyên: - `houses`: chứa vị trí của các ngôi nhà. - `heaters`: chứa vị trí của các lò sưởi. 
Hàm trả về một số nguyên là bán kính r nhỏ nhất cần thiết. 
 
Lưu ý: - Số lượng nhà và lò sưởi là một số không âm và không vượt quá 25000. - Vị trí của các ngôi nhà và lò sưởi là một số không âm và không vượt quá 10^9.

```python
def findRadius(H, T):
    T.sort()
    r = 0
    m = len(T)
    
    for h in H:
        low, high = 0, m - 1
        idx = m  
        
        while low <= high:
            mid = (low + high) // 2
            if T[mid] >= h:
                idx = mid     
                high = mid - 1
            else:
                low = mid + 1

        d_right = T[idx] - h if idx < m else float('inf')

        d_left = h - T[idx - 1] if idx > 0 else float('inf')

        r = max(r, min(d_left, d_right))
        
    return r

houses = list(map(int, input().split()))
heaters = list(map(int, input().split()))

print(findRadius(houses, heaters))
```

#### W9A13
Viết hàm power(base, exp) tính base^exp bằng đệ quy chia để trị.  
 
Yêu cầu:  - Không dùng vòng lặp  - Không dùng hàm có sẵn

```python
def power(base, exp):
    if exp == 0:
        return 1
    if exp < 0:
        return 1 / power(base, -exp)
    
    half = power(base, exp // 2)

    if exp % 2 == 0:
        return half * half
    else:
        return base * half * half
    
base, exp = map(int, input().split())

print(power(base, exp))
```

#### W9A14
VIế ́ t hàm tìm số xuất hiện nhiều hơn x lần trong một dãy số (x là số nguyên dương nhập 
từ bàn phím). Nếu không có số nào thì in ra "INVALID" 

```python
def main(a, x):
    dic = {}
    A = []
    
    for i in a:
        if i not in dic:
            dic[i] = 1
        else:
            dic[i] += 1

    for i in dic:
        if dic[i] == x:
            A.append(i)

    if len(A) == 0:
        return 'INVALID'
    
    return A

a = list(map(int,input().split()))
x = int(input())

print(main(a, x))
```

#### W9A15
VIết haàm kiểm tra dãy số nguyên vào từ bàn phím có phải dãy tăng nghiêm ngặt (giảm 
nghiêm ngặt) hay không?

```python
def main(a):
    tang = True
    giam = True
    n = len(a)

    if n < 2:
        return None
    
    for i in range(n - 1):
        if a[i] < a[i+1]:
            giam = False
        if a[i] > a[i+1]:
            tang = False

    if tang == True:
        return 'tang'
    elif giam == True:
        return 'giam'
    else:
        return 'None'
    
a = list(map(int, input().split()))

print(main(a))
```
