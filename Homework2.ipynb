#### Bài 3
Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là 
độ dài ba cạnh hay không, nếu  không thì in  ra màn  hình  'Khong la ba  canh 
cua mot tam  giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, 
Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn 
hình. 

```python
a = float(input())
b = float(input())
c = float(input())
if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print('tam giác đều')
    elif a == b or a == c or b == c:
        print('tam giác cân')
    else:
        if a**2 == b**2 + c**2 or b**2 == a**2 + c**2 or c**2 == a**2 + b**2:
            print('tam giác vuông')
        else:
            print('tam giác thường')
else:
    print('không là 3 cạnh của một tam giác')
```

#### Bài 5
Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0) 

```python
a, b, c = map(float, input().split())
delta = b**2 - 4*a*c
if delta < 0:
    print('phương trình vô nghiệm')
else :
    x1 = (-b + delta**0.5) / (2*a)
    x2 = (-b - delta**0.5) / (2*a)
    print(f'x1 = {x1}, x2 = {x2}')
```

#### Bài 8
Giải hệ phương trình tuyến tính: 

ax + by = m
 
cx + dy = n 

```python
a, b, m, c, d, n = map(float, input().split())
det = a * d - b * c
if det == 0:
    if a * n == m * c and b * n == m * d:
        print('hệ phương trình có vô số nghiệm')
    else:   
        print('hệ phương trình vô nghiệm')
else:
    x = (m * d - b * n) / det
    y = (a * n - m * c) / det
    print(f'x = {x}, y = {y}')
```

#### Bài 10
Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một 
điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
I = input().split()
r = float(input())
A = input().split()
if ( A[0] - I[0]) ** 2 + (A[1] - I[1]) ** 2 == r ** 2:
    print('điểm A thuộc đường tròn')
```
