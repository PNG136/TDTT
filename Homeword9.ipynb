#### W9A16
Cho hai số nguyên dương, n và m (0 < n, m <= 20 ). .Bạn hãy in ra tất cả các chuỗi nhị phân có 
n số 0 và m số 1. In theo thứ tự tăng dần trong từ điển 

```python
def main(a, n, m):
    a = bin(a)[2::]
    for i in a:
        if int(i) == 0:
            m -= 1
        else:
            n -= 1

    if m == 0 and n == 0:
        return True
    
n, m = map(int, input().split())
k = 0

for i in range(m):
    k += 2 ** (m + n - i)

for i in range(k + 1):
    if main(i, n, m):
        print(bin(i)[2::], end = ' ')
```

#### W9A17
Viết hàm tìm dãy con tăng nghiêm ngặt dài nhất trong dãy đầu vào từ bàn phím

```python
def main(a):
    r = []
    
    for v in a:
        i, j = 0, len(r)

        while i < j:
            m = (i + j) // 2

            if r[m] < v:
                i = m + 1               
            else:
                j = m

        if i == len(r):
            r.append(v)
        else:
            r[i] = v

    return len(r)

A = list(map(int, input().split()))

print(main(A))
```

#### W9A20
Thực hiện chuẩn hóa chuỗi kí tự: 
 
a. bỏ khoảng trắng đầu cuối 
 
b. các khoảng trắng giữa chỉ còn 1 dấu cách 
 
c. viết hoa chữ cái đầu mỗi từ 
KHÔNG DÙNG CÁC HÀM SẴN CỐ 

```python
def clean(s):
    res = ""
    in_word = False 
    
    for c in s:
        if c == ' ' or c == '\t':
            if in_word: 
                res += ' ' 
                in_word = False
        else:
            o = ord(c)

            if not in_word:
                if 97 <= o <= 122:
                    c = chr(o - 32)
                in_word = True
            else:
                if 65 <= o <= 90:
                    c = chr(o + 32)
            
            res += c

    if len(res) > 0 and res[-1] == ' ':
        res = res[:-1]
        
    return res

s = input()

print(clean(s))
```

#### W9A21
Tìm phần tử lớn thứ k trong dãy n phần tử bằng thuật toán tốt nhất có thể – Tránh dùng sắp xếp 
(dùng phân hoạch của quicksort) 

```python
def part(A, L, R):
    piv = A[R]
    i = L
    for j in range(L, R):
        if A[j] < piv:
            A[i], A[j] = A[j], A[i]
            i += 1
    A[i], A[R] = A[R], A[i]
    return i

def select(A, L, R, t):
    if L == R:
        return A[L]

    p = part(A, L, R)

    if p == t:
        return A[p]

    elif p > t:
        return select(A, L, p - 1, t)

    else:
        return select(A, p + 1, R, t)

def main(A, k):
    n = len(A)

    return select(A, 0, n - 1, n - k)

A = list(map(float, input().split()))
k = int(input())

print(main(A, k))
```

#### W9A22
Hai chuỗi là anagram nếu chúng có cùng ký tự và số lần xuất hiện. viết CT kiểm tra xem 2 chuỗi 
đầu vào có phải là anagram của nhau không 

```python
def anagram(s, t):
    if len(s) != len(t):
        return False
        
    d = {}

    for c in s:
        d[c] = d.get(c, 0) + 1

    for c in t:
        if c not in d or d[c] == 0:
            return False
        d[c] -= 1
        
    return True

s = list(map(float, input().split()))
t = list(map(float, input().split()))

print(anagram(s, t))
```

#### W9A23
Cho số nguyên rất lớn (string), và một số nguyên dương k. Tìm số có k chữ số lớn nhất có thể 
từ số ban đầu bằng cách xóa bớt đi một số chữ số nào đó mà không thay đổi vị trí của bất kì 
chữ số nào. (VD: 78934340999, k = 7 => 9440999 là số lớn nhất có 7 chữ số cần tìm) 

```python
def max_num(s, k):
    st = []

    rem = len(s) - k 
    
    for c in s:
        while st and st[-1] < c and rem > 0:
            st.pop()
            rem -= 1
        st.append(c)

    return "".join(st[:k])

s = input()
k = int(input())

print(max_num(s, k))
```

#### W9A24
Nén chuối kí tự bằng cách đếm số lần xuất hiện của các kí tự liên tiếp nhau và ghi lại ngay sau 
kí tự đó. VD aaaaaafdhhhhh =>a6f1d1h5 

```python
def main(a):
    n = len(a)
    k = ''
    m = 0

    if n == 0:
        return None
    
    if n == 1:
        return a + '1'

    while m < n:
        x = 1 

        while m + 1 < n and a[m] == a[m + 1]:
            x += 1
            m += 1 

        k += a[m] + str(x)
        m += 1 

    return k

a = input("Nhập chuỗi: ")
print(main(a))

        
```

#### W9A25
Giả sử mật khẩu mạnh là một chuỗi kí tự dài ít nhất 8 kí tự gồm ít nhất 1 chữ cái thường, 1 chữ

cái hoa, 1 chữ số và 1 trong các kí tự đặc biệt !, @, #, $. Nhập vào một chuỗi kĩ tự, kiểm tra xem

nó có phải là mật khẩu mạnh không.

```python
def main(A):
    n = len(A)

    if n < 8:
        return False
    
    a = False
    b = False
    c = False
    d = False

    B = ['!', '@', '#', '$']

    for i in A:
        if i.isupper():
            a = True
        elif i.islower():
            b = True
        elif i.isdigit():
            c = True
        elif i in B:
            d = True

    return a and b and c and d

A = input()

if main(A):
    print("Mật khẩu này MẠNH!")
else:
    print("Mật khẩu này YẾU!")
```
