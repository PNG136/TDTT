#### Bài 4
Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím 

```python
h = int(input())
for i in range(h-1):
    print(" " * (h - i - 1) + "*" + " " * (2 * i - 1) + "*" if i > 0 else " " * (h - i - 1) + "*")
print("*" * (2 * h - 1))
```

#### Bài 8
Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b 

```python
a = int(input())
b = int(input())
while a != b:
    if a > b:
        a -= b
    else:
        b -= a
for i in range(1, a+1):
    if a % i == 0:
        print(i, end=' ')
```

#### Bài 9
 Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay 
không?

```python
a = int(input())
n = 0
while (n+1) * (n//2 +1) // 2 < a:
    n += 1
    if (n+1) * (n//2 +1) // 2 == a:
        print('a là số chính phương')
if (n+1) * (n//2 +1) // 2 != a:
    print('a không phải là số chính phương')
```

#### Bài 13
Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho 
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 

```python
A = int(input())
while A < 1:
    A = int(input())
k = 1
s = 1
while s <= A:
    s += 1/(k+1)
    k += 1
print(k)
```

#### Bài 14
Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập 
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
A = int(input())
max = A
min = A
while A != -1:
    if A > max:
        max = A
    if A < min:
        min = A
    A = int(input())
print(max)
print(min)
```

#### Bài 18
Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay 
không 

```python
n = int(input())
if n % 3 != 0:
    print('no')
if n == 1:
    print('yes')
else:
    while n % 3 == 0:
        n //= 3
    if n  == 1:
        print('yes')
    else:
        print('no')
```

#### Bài 19
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số 
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong 
dãy số fibonacci lớn nhất nhưng không vượt quá A 

```python
A = int(input())
a = 0
b = 1
n = 1
while A > n:
    a = b
    b = n
    n = a + b
print(n-a)
```
