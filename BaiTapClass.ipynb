#### W8A1
[Cylinder] 
Xây dựng lớp Cylinder (hình trụ) gồm: bán kính và chiều cao 
(kiểu số thực) với hai phương thức:tính diện tích bề mặt (diện 
tích toàn phần) và thể tích của hình trụ 
Đầu vào  
Đâu vào từ bàn phím gồm một dòng duy nhất chứa bán kính và 
chiều cao của hình trụ, cách nhau bởi một dấu cách. 
 
Đầu ra 
In ra màn hình diện tích bề mặt và thể tích của hình trụ (sử 
dụng lại những hàm vừa viết, π=3.14).  
 
(Kết quả làm tròn đến 2 chữ số sau dấu phẩy). 

```python
class Cylinder:
    def __init__(self, r, h):
        self.r = r
        self.h = h

    def S(self):
        return 2 * self.r * (self.r + self.h) * 3.14

    def V(self):
        return self.r * self.r * self.h * 3.14

r, h = map(float, input().split())

cylinder = Cylinder(r, h)

print(f'S = {cylinder.S():.2f}')
print(f'V = {cylinder.V():.2f}')
```

[Date] 
Xây dựng lớp Date biểu diễn ngày tháng năm với hai phương 
thức kiểm tra xem một ngày tháng nào đó có hợp lệ hay không 
và phương thức tính ngày kế tiếp của ngày tháng năm nào đó 
nếu nó là hợp lệ 
 
Nhập vào một ngày tháng năm, (ngày tháng năm biểu diễn 
dưới dạng dd/mm/yyyy). Nếu ngày tháng năm không tồn tại, in 
ra dòng chữ INVALID. Nếu là hợp lệ in ra ngày kế tiếp của nó. 
 
Đầu vào  
Đầu vào từ bàn phim chứa dữ liệu ngày, tháng, năm dưới dạng 
dd/mm/yyyy 
. 
Đầu ra 
In ra màn hình ngày kế tiếp của ngày vừa nhập dưới dạng 
dd/mm/yyyy 
. Nếu ngày tháng không tồn tại, in ra INVALID 

```python
class Date:
    def __init__(self, d, m, y):
        self.d = d
        self.m = m
        self.y = y

    def is_valid(self):
        maxday = -1

        match self.m:

            case 1|3|5|7|8|10|12:
                maxday = 31

            case 4|6|9|11:
                maxday = 30

            case 2:
                if (self.y % 4 == 0 and self.y % 100 != 0) or (self.y % 400 == 0):
                    maxday = 29

                else:
                    maxday = 28
            
        return 1 <= self.d <= maxday and self.y > 0
    
    def next_day(self):
        nextday = Date(self.d + 1, self.m, self.y)

        if nextday.is_valid():
            return nextday
        
        nextday = Date(1, self.m + 1, self.y)

        if nextday.is_valid():
            return nextday
        
        nextday = Date(1, 1, self.y + 1)

        if nextday.is_valid():
            return nextday
        
d, m, y = map(int, input().split('/'))

date = Date(d, m, y)

if not date.is_valid():
    print('INVALID')

else:
    nextday = date.next_day()

    print(f'{nextday.d:02d}/{nextday.m:02d}/{nextday.y:04d}')
```

#### W8A4
[Calculator] 
Xây dựng lớp Calculator (máy tính bỏ túi) với các phương thức 
để thực hiện các phép toán 1. add (cộng), 2. subtract(trừ), 3. 
product (nhân) và 4. divide(chia), 5. power(hàm mũ), 6. mod 
(lấy phần dư) và phương thức 7. set_numbers(cập nhật 2 số 
mới) nếu người dùng muốn tính toán trên 2 số khác. 
Khởi tạo lớp là hai thuộc tính gồm 2 số cần tính, Mỗi phương 
thức thực hiện một phép tính và trả về kết quả tương ứng 
Tạo menu và cho người dùng tính toán tới khi gõ "yes", "YES" 
hoặc "Yes" cho câu hỏi "Bạn có muốn thoát?" mới dừng.

```python
class Calculator:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

    def add(self):
        return self.num1 + self.num2

    def subtract(self):
        return self.num1 - self.num2

    def product(self):
        return self.num1 * self.num2

    def divide(self):
        if self.num2 == 0:
            return "Lỗi: Không thể chia cho 0!"
        return self.num1 / self.num2

    def power(self):
        return self.num1 ** self.num2

    def mod(self):
        if self.num2 == 0:
            return "Lỗi: Không thể chia lấy dư cho 0!"
        return self.num1 % self.num2

    def set_numbers(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

def show_menu():
    print("\n" + "="*30)
    print("      MÁY TÍNH BỎ TÚI")
    print("="*30)
    print("1. Phép cộng (+)")
    print("2. Phép trừ (-)")
    print("3. Phép nhân (*)")
    print("4. Phép chia (/)")
    print("5. Hàm mũ (**)")
    print("6. Chia lấy dư (mod)")
    print("7. Đổi 2 số mới")
    print("="*30)

print("Khởi tạo máy tính:")
n1 = float(input("Nhập số thứ nhất: "))
n2 = float(input("Nhập số thứ hai: "))

calc = Calculator(n1, n2)

while True:
    print(f"\n[Hai số hiện tại là: {calc.num1} và {calc.num2}]")
    show_menu()
    
    choice = input("Chọn phép toán (1-7): ").strip()
    
    match choice:
        case '1':
            print(f"Kết quả: {calc.num1} + {calc.num2} = {calc.add()}")
        case '2':
            print(f"Kết quả: {calc.num1} - {calc.num2} = {calc.subtract()}")
        case '3':
            print(f"Kết quả: {calc.num1} * {calc.num2} = {calc.product()}")
        case '4':
            print(f"Kết quả: {calc.num1} / {calc.num2} = {calc.divide()}")
        case '5':
            print(f"Kết quả: {calc.num1} ^ {calc.num2} = {calc.power()}")
        case '6':
            print(f"Kết quả: {calc.num1} % {calc.num2} = {calc.mod()}")
        case '7':
            new_n1 = float(input("Nhập số thứ nhất mới: "))
            new_n2 = float(input("Nhập số thứ hai mới: "))
            calc.set_numbers(new_n1, new_n2)
            print("Đã cập nhật 2 số mới thành công!")
        case _:
            print("Lựa chọn không hợp lệ, vui lòng chọn từ 1 đến 7.")
            continue 

    exit_choice = input("\nBạn có muốn thoát? (Gõ 'yes'/'YES'/'Yes' để thoát): ").strip()

    if exit_choice.lower() == 'yes':
        print("Cảm ơn bạn đã sử dụng máy tính!")
        break
```

#### W8A5
[ShoppingCart] 
Xây dựng lớp ShoppingCart (giỏ hàng) để quản lý các sản 
phẩm của khách. Dùng danh sách để lưu các sản phẩm, mỗi sản phẩm là một từ điển có name (tên) và price(giá).  
Với các phương thức sau: 
1. Khởi tạo với danh sách rỗng 
2. Thêm vào giỏ hàng một sản phẩm mới 
3. Xóa khỏi giỏ một sản phẩm nào đó (tìm theo tên để xóa) 
4. Kiểm tra xem giỏ hàng có rỗng không 
5. Tính tổng tiền giỏ hàng 
6. Hiển thị toàn bộ sản phẩm trong giỏ 
7. Xóa toàn bộ giỏ hàng 
 
Tạo menu cho người dùng nhập lựa chọn truy vấn từ 2 tới 7. 
Chương trình dừng khi người dùng nhập "yes" cho câu hỏi "Có 
muốn thoát khỏi chương trình không?" 

```python
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, name, price):
        product = {"name": name, "price": price}
        self.items.append(product)
        print(f"Đã thêm sản phẩm '{name}' với giá {price:,}đ vào giỏ hàng.")

    def remove_item(self, name):
        for item in self.items:
            if item["name"].lower() == name.lower():
                self.items.remove(item)
                print(f"Đã xóa sản phẩm '{item['name']}' khỏi giỏ hàng.")
                return True
        print(f"Không tìm thấy sản phẩm nào có tên '{name}' trong giỏ hàng.")
        return False

    def is_empty(self):
        return len(self.items) == 0

    def get_total(self):
        total = sum(item["price"] for item in self.items)
        return total

    def display_cart(self):
        if self.is_empty():
            print("Giỏ hàng của bạn đang trống.")
            return

        print("\n--- DANH SÁCH SẢN PHẨM TRONG GIỎ ---")
        for i, item in enumerate(self.items, 1):
            print(f"{i}. Tên: {item['name']} | Giá: {item['price']:,}đ")
        print(f"--> Tổng cộng: {self.get_total(),}đ")

    def clear_cart(self):
        self.items.clear()
        print("Đã xóa toàn bộ sản phẩm, giỏ hàng hiện tại trống rỗng.")


def show_menu():
    print("\n" + "="*35)
    print("         MENU GIỎ HÀNG")
    print("="*35)
    print("2. Thêm sản phẩm vào giỏ hàng")
    print("3. Xóa sản phẩm theo tên")
    print("4. Kiểm tra giỏ hàng có rỗng không")
    print("5. Tính tổng tiền giỏ hàng")
    print("6. Hiển thị toàn bộ giỏ hàng")
    print("7. Xóa sạch giỏ hàng")
    print("="*35)

cart = ShoppingCart()

while True:
    show_menu()
    choice = input("Nhập lựa chọn của bạn (2-7): ").strip()

    match choice:
        case '2':
            name = input("Nhập tên sản phẩm: ").strip()

            try:
                price = float(input("Nhập giá sản phẩm: "))
                if price < 0:
                    print("Giá tiền không thể âm!")
                else:
                    cart.add_item(name, price)
            except ValueError:
                print("Lỗi: Giá tiền phải là một số!")

        case '3':
            name_to_remove = input("Nhập chính xác tên sản phẩm cần xóa: ").strip()
            cart.remove_item(name_to_remove)

        case '4':
            if cart.is_empty():
                print("Giỏ hàng hiện đang RỖNG.")
            else:
                print(f"Giỏ hàng KHÔNG rỗng (đang có {len(cart.items)} sản phẩm).")

        case '5':
            print(f"Tổng tiền các sản phẩm trong giỏ: {cart.get_total(),}đ")

        case '6':
            cart.display_cart()

        case '7':
            confirm = input("Bạn có chắc chắn muốn xóa TOÀN BỘ giỏ hàng? (y/n): ").strip().lower()
            if confirm == 'y':
                cart.clear_cart()
            else:
                print("Đã hủy thao tác xóa toàn bộ.")

        case _:
            print("Lựa chọn không hợp lệ! Vui lòng chọn từ số 2 tới số 7.")
            continue

    exit_choice = input("\nCó muốn thoát khỏi chương trình không? (gõ 'yes' để thoát): ").strip().lower()
    if exit_choice == 'yes':
        print("Cảm ơn bạn đã sử dụng dịch vụ giỏ hàng!")
        break
```

Xây dựng lớp Rectangle(w, h) với các phương thức: 
 
area() trả về diện tích. 
 
perimeter() trả về chu vi. 
 
scale(k) phóng to hình chữ nhật theo hệ số k (nhân cả w và h), 
không trả về gì. 
 
Input 
w h k (số thực dương) 
 
Output 
area perimeter sau khi phóng to (làm tròn 2 chữ số thập phân). 

```python
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h
    
    def perimeter(self):
        return 2 * (self.w + self.h)
    
    def scale(self, k):
        self.w *= 2
        self.h *= 2

w, h, k = map(float, input().split())

rec = Rectangle(w, h)

rec.scale(k)

print(f'{rec.area():.2f}')
print(f'{rec.perimeter():.2f}')
```

#### W8A7
Yêu cầu 
Lớp Fraction(num, den) (tử, mẫu nguyên, mẫu ≠ 0) với: 
 
Rút gọn phân số trong __init__ (quy ước mẫu > 0). 
 
Các phương thức add/sub/mul/div trả về phân số đã rút gọn. 
 
__str__() in dạng a/b. 
 
Input 
a b op c d (tương ứng a/b (op) c/d, op thuộc {+,-,*,/}) 
 
Output 
Phân số kết quả đã rút gọn 

```python
class Fraction:
    def __init__(self, num, den):
        if den == 0:
            raise ValueError()

        if den < 0:
            num = -num
            den = -den

        common_gcd = self._gcd(abs(num), abs(den))
        self.num = num // common_gcd
        self.den = den // common_gcd

    def _gcd(self, x, y):
        while y != 0:
            x, y = y, x % y
        return x if x != 0 else 1

    def add(self, other):
        new_num = self.num * other.den + other.num * self.den
        new_den = self.den * other.den
        return Fraction(new_num, new_den)

    def sub(self, other):
        new_num = self.num * other.den - other.num * self.den
        new_den = self.den * other.den
        return Fraction(new_num, new_den)

    def mul(self, other):
        new_num = self.num * other.num
        new_den = self.den * other.den
        return Fraction(new_num, new_den)

    def div(self, other):
        if other.num == 0:
            raise ZeroDivisionError()
        new_num = self.num * other.den
        new_den = self.den * other.num
        return Fraction(new_num, new_den)

    def __str__(self):

        if self.den == 1:
            return f"{self.num}"
        return f"{self.num}/{self.den}"

try:
    line = input().split()
    if len(line) == 5:
        a = int(line[0])
        b = int(line[1])
        op = line[2]
        c = int(line[3])
        d = int(line[4])

        f1 = Fraction(a, b)
        f2 = Fraction(c, d)
        result = None

        match op:
            case '+': result = f1.add(f2)
            case '-': result = f1.sub(f2)
            case '*': result = f1.mul(f2)
            case '/': result = f1.div(f2)
            case _: print("Phép toán không hợp lệ!")

        if result:
            print(result)

except Exception as e:
    print(f"Lỗi: {e}")
```

#### W8A8
Yêu cầu 
Lớp BankAccount(owner, balance) với: 
 
deposit(x), withdraw(x); nếu rút quá số dư thì bỏ qua giao dịch. 
 
__str__() in owner: balance. 
 
Input 
Dòng 1: owner initial_balance 
Dòng 2: n (số lệnh) 
Tiếp theo n dòng, mỗi dòng: DEPOSIT x hoặc WITHDRAW x. 
 
Output 
Số dư cuối cùng (in số thực, 2 chữ số). 

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance) 

    def deposit(self, x):
        if x > 0:
            self.balance += x

    def withdraw(self, x):
        if 0 < x <= self.balance:
            self.balance -= x

    def __str__(self):
        return f"{self.owner}: {self.balance:.2f}"

try:
    line1 = input().split()
    owner = line1[0]
    initial_balance = float(line1[1])

    account = BankAccount(owner, initial_balance)

    n = int(input())

    for _ in range(n):
        command = input().split()
        action = command[0].upper()  
        amount = float(command[1])

        match action:
            case "DEPOSIT":
                account.deposit(amount)
            case "WITHDRAW":
                account.withdraw(amount)

    print(account)

except Exception as e:
    print(f"Lỗi định dạng đầu vào: {e}")
```

#### W8A9
Yêu cầu 
Lớp Student(name) với: 
 
add_score(subject, score) 
 
avg() trả về điểm trung bình. 
 
rank() trả về "Excellent" nếu avg ≥ 8, "Good" nếu 6.5–<8, 
"Average" nếu 5–<6.5, ngược lại "Poor". 
 
Input 
Dòng 1: name 
Dòng 2: n 
n dòng tiếp: subject score (score là số thực 0–10) 
 
Output 
name avg rank (avg 2 chữ số).

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []

    def add_score(self, subject, score):
        self.scores.append(float(score))

    def avg(self):
        if not self.scores:
            return 0.0
        return sum(self.scores) / len(self.scores)

    def rank(self):
        average = self.avg()
        if average >= 8.0:
            return "Excellent"
        elif average >= 6.5:
            return "Good"
        elif average >= 5.0:
            return "Average"
        else:
            return "Poor"

try:
    name = input().strip()

    student = Student(name)

    n = int(input())

    for _ in range(n):
        line = input().split()

        score = float(line[-1])
        subject = " ".join(line[:-1])
        
        student.add_score(subject, score)

    final_avg = student.avg()
    final_rank = student.rank()

    print(f"{student.name} {final_avg:.2f} {final_rank}")

except Exception as e:
    print(f"Lỗi định dạng dữ liệu đầu vào: {e}")
```

#### W8A10
Yêu cầu 
 
Lớp Book(title, author, year). 
 
Lớp Library() với: 
 
add(book) 
 
count_by_author(name) 
 
find_by_year(y) trả về số sách xuất bản năm y. 
 
Input 
Dòng 1: n (số thao tác) 
n dòng: 
 
ADD title;author;year 
 
COUNT author (đếm theo tác giả) 
 
COUNTYEAR year (đếm theo năm) 
 
Output 
Mỗi lệnh COUNT/COUNTYEAR in một dòng kết quả. 

```python
class Book:
    def __init__(self, title, author, year):
        self.title = title.strip()
        self.author = author.strip()
        self.year = int(year)


class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def count_by_author(self, name):
        count = 0
        for book in self.books:
            if book.author.lower() == name.strip().lower():
                count += 1
        return count

    def find_by_year(self, y):
        count = 0
        for book in self.books:
            if book.year == int(y):
                count += 1
        return count

try:
    n = int(input())

    library = Library()

    for _ in range(n):
        line = input().strip()
        if not line:
            continue

        parts = line.split(" ", 1)
        command = parts[0].upper()
        
        match command:
            case "ADD":
                details = parts[1].split(";")
                title = details[0]
                author = details[1]
                year = details[2]

                new_book = Book(title, author, year)
                library.add(new_book)
                
            case "COUNT":
                author_name = parts[1]
                print(library.count_by_author(author_name))
                
            case "COUNTYEAR":
                target_year = parts[1]
                print(library.find_by_year(target_year))

except Exception as e:
    print(f"Lỗi định dạng đầu vào: {e}")
```

In ra góc giữa hai mặt phẳng, chính xác tới hai chữ số thập 
phân. 

```python
import math

def dot(v1, v2):
    return v1[0] * v2[0] + v1[1] * v2[1] + v1[2] * v2[2]

def cross(v1, v2):
    return [
        v1[1] * v2[2] - v1[2] * v2[1],
        v1[2] * v2[0] - v1[0] * v2[2],
        v1[0] * v2[1] - v1[1] * v2[0]
    ]

def sub(p1, p2):
    return [p1[0] - p2[0], p1[1] - p2[1], p1[2] - p2[2]]

def mag(v):
    return math.sqrt(dot(v, v))

A = list(map(float, input().split()))
B = list(map(float, input().split()))
C = list(map(float, input().split()))
D = list(map(float, input().split()))

ab = sub(B, A)
ac = sub(C, A)
bc = sub(C, B)
bd = sub(D, B)

n1 = cross(ab, ac)
n2 = cross(bc, bd)

m1 = mag(n1)
m2 = mag(n2)

if m1 == 0 or m2 == 0:
    print(f"{0.00:.2f}")
else:
    cos_phi = abs(dot(n1, n2)) / (m1 * m2)
    cos_phi = min(1.0, max(-1.0, cos_phi))
    phi = math.degrees(math.acos(cos_phi))
    print(f"{phi:.2f}")
```

[Complex] 
 
Khai báo kiểu dữ liệu biểu diễn số phức \(ComplexNumber\) 
bao gồm \(2\) biến \(realPart\) và \(imaginaryPart\) kiểu 
\(double\) biểu diễn phần thực và phần ảo của một số phức. 
 
Viết hàm khởi tạo ComplexNumber(double initRealPart, double 
initImaginaryPart) để khởi tạo các giá trị phần thực và phần ảo 
của một số phức và hàm ComplexNumber 
multiply(complexNumber other) trả về tích của số phức hiện tại 
và một số phức khác. 
 
Đầu vào 
 
Đầu vào từ bàn phím gồm hai dòng, mỗi dòng gồm hai số thực 
cách nhau bởi một dấu cách biểu diễn phần thực và phần ảo 
của một số phức. Các số trong đầu vào có giá trị tuyệt đối 
không vượt quá \(100\). 
 
Đầu ra 
 
In ra màn hình hai số biểu diễn phần thực và phần ảo của tích 

```python
class ComplexNumber:
    def __init__(self, r, i):
        self.r = r
        self.i = i

    def multiply(self, o):
        r = self.r * o.r - self.i * o.i
        i = self.r * o.i + self.i * o.r
        return ComplexNumber(r, i)

r1, i1 = map(float, input().split())
r2, i2 = map(float, input().split())

c1 = ComplexNumber(r1, i1)
c2 = ComplexNumber(r2, i2)
res = c1.multiply(c2)

print(f"{res.r:.2f} {res.i:.2f}")
```

#### W8A13
[Triangle] 
 
Khai báo cấu trúc dữ liệu tam giác \(Triangle\) chứa \(3\) biến 
\(egde1,\,edge2,\,edge3\) kiểu \(int\) là độ dài \(3\) cạnh của 
tam giác đó. 
 
Viết hàm khởi tạo Triangle(int initEdge1,int initEdge2, int 
initEdge3) để khởi tạo độ dài các cạnh của tam giác và hàm 
double getArea() trả về diện tích của tam giác đó (nếu tam giác 
không hợp lệ, trả về \(-1\)). 
 
Hoàn thiện chương trình dưới đây nhập vào \(3\) cạnh của một 
tam giác và in ra diện tích của tam giác đó. 
 
Đầu vào 
 
Một dòng duy nhất từ bàn phím chứa \(3\) số nguyên là độ dài 
\(3\) cạnh của tam giác 
 
Đầu ra 
 
In ra màn hình diện tích tam giác (làm tròn đến\(2\) chữ số thập 
phân sau dấu phẩy). In ra \(invalid\) nếu \(3\) số nhập vào 
không phải là \(3\) cạnh của một tam giác 

```python
class Triangle:
    def __init__(self, e1, e2, e3):
        self.e1 = e1
        self.e2 = e2
        self.e3 = e3

    def getArea(self):
        a, b, c = self.e1, self.e2, self.e3
        if a + b <= c or a + c <= b or b + c <= a or a <= 0 or b <= 0 or c <= 0:
            return -1.0
        p = (a + b + c) / 2.0
        return (p * (p - a) * (p - b) * (p - c)) ** 0.5

e1, e2, e3 = map(int, input().split())
t = Triangle(e1, e2, e3)
res = t.getArea()

if res == -1.0:
    print("invalid")
else:
    print(f"{res:.2f}")
```

#### W8A14
Yêu cầu 
Xây dựng lớp Animal với phương thức speak(). 
Xây dựng hai lớp kế thừa Dog và Cat, mỗi lớp ghi đè speak() 
để trả về âm thanh tương ứng: - Dog -> Woof - Cat -> Meow 
 
Input 
Một dòng chứa tên lớp: Dog hoặc Cat.
Output 
In ra kết quả của speak(). 

```python
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof"

class Cat(Animal):
    def speak(self):
        return "Meow"

t = input().strip()
if t == "Dog":
    print(Dog().speak())
elif t == "Cat":
    print(Cat().speak())
```

#### W8A15
Yêu cầu 
Xây dựng lớp Employee(name, salary). 
Xây dựng lớp Manager(name, salary, bonus) kế thừa từ 
Employee. 
Cả hai lớp đều có phương thức income(): - Employee: trả về salary - Manager: trả về salary + bonus 
Input 
Dòng 1: loại đối tượng (Employee hoặc Manager) 
Dòng 2: name salary [bonus] 
Output 
In ra tổng thu nhập của đối tượng.

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def income(self):
        return self.salary

class Manager(Employee):
    def __init__(self, name, salary, bonus):
        super().__init__(name, salary)
        self.bonus = bonus

    def income(self):
        return self.salary + self.bonus

t = input().strip()
d = input().split()

if t == "Employee":
    name = d[0]
    salary = float(d[1])
    obj = Employee(name, salary)
elif t == "Manager":
    name = d[0]
    salary = float(d[1])
    bonus = float(d[2])
    obj = Manager(name, salary, bonus)

print(f"{obj.income():.2f}")
```
