#### W4A6
"[SumPrimesInRange] 

Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn 
[a, b]. Mỗi dòng là một bộ test (a b)."

```python
def main(a):
    if a < 2:
        return -1
    if a == 2:
        return 1
    if a % 2 == 0:
        return -1
    for i in range(3, a//2 + 1, 2):
        if a % i == 0:
            return -1
    return 1
while True:
    a, b = map(int, input().split())
    if b >= a:
        break
n = 0
for i in range(a , b+1):
    k = main(i)
    if k == 1:
        n += i
print(n)
```

Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n." 


```python
while True:
    n = int(input())
    if n >= 2:
        break
k = 2
a = 1
while n != 1:
    if n % k == 0:
        n //= k
        a = k
        continue
    k += 1
print(a)
```

[ReverseAddPalindrome]

đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 

```python
n = int(input())
a = 0
if n < 10:
    print(0, n)
else:
    while str(n)[0:len(str(n))//2] != str(n)[len(str(n))//2:][::-1] and a < 1000:
        n += int(str(n)[::-1])
        a += 1
        if a == 1000000000:
            print("Không tìm thấy số đối xứng")
            break
    else:
        print(a, n)
```

#### W4A10
"[LongestCollatzUpToN] 

Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất 
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số 
nhỏ nhất." 

```python
n = int(input())
h = 0
y = 1
x = 1
while 1 <= x <= n:
    k = 0
    a = x
    while x != 1:
        if x % 2 == 0:
            x //= 2
            k += 1
            continue
        else:
            x = 3*x + 1
            k += 1
            continue
    if h <= k:
        if y >= a:
            y = a
            h = k
        else:
            y = a
            h = k
    x = a + 1
print(y, h)
```

#### W4A11
[EvenDivisors] 

Viết chương trình đếm số lượng ước số chẵn của số nguyên dương n với n < 10^6. 

```python
def main(n):
    k = 3
    a = 1
    b = 0
    while n % 2 == 0:
        n //= 2
        b += 1
    while n != 1:
        if n % k == 0:
            n //= k
            a += 1
            continue
        k += 1
    return a * b
n = int(input())
if n == 1:
    print(0)
else: 
    print(main(n))
```

Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không. 
Nếu có, in ra màn hình "true", ngược lại, in ra "false"

```python
n = int(input())
m = int(input())
k = 0
h = 0
for i in range(1,n//2 + 1):
    if n % i == 0:
        k += i
if k == m:
    for i in range(1,m//2 + 1):
        if m % i == 0:
            h += i
    if h == n:
        print('true')
    else: 
        print('false')
else: 
    print('false')
```

#### W4A14
[GCD] 

Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình 
tìm ước chung lớn nhất của hai số m và n.

```python
m = int(input())
n = int(input())
while m != n:
    if m > n:
        m -= n
    else:
        n -= m
print(m)
```
