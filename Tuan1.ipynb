#### BÀI 2
Một khu đất có chiều rộng a, chiều dài b. Người ta xây một khu vui chơi hình tròn tại vị trí trung tâm như hình vẽ. Nhập vào a,b. Hãy tính phần diện tích còn lại để troòng cây là bao. Biết pi = 3.14, lấy kết quả đến 2 chữ số thập phân.

```python
a = float(input('chiều rộng: '))
b = float(input('chiều dài: '))
print('diện tích để trồng cây là:', round(a * b - min(a, b) ** 2 * 3.14, 2))
```

#### Bài 6
Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không, nếu đúng in ra diện tích tam giác đó, in ra với 1 chữ số thập phân, nếu không phải 3 cạnh tam giác thì in ra "không phải 3 cạnh của tam giác"

```python
a = float(input())
b = float(input())
c = float(input())
if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    print('diện tích tam giác là:', round((p * (p - a) * (p - b) * (p - c)) ** 0.5, 1))
else:
    print("không phải 3 cạnh của tam giác")
```

#### Bài 8
Nhập vào họ tên chủ hộ và chỉ số điện kế tháng trước và chỉ số điện kế tháng này, 
tính tiền điện tháng này cho hộ biết:
- Từ 0 – 50kWh: 1.984 đồng/kWh 
- Bậc 2: Từ 51 – 100 kWh: 2.050 đồng/kWh  
- Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh  
- Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh  
- Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh  
- Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh 

```python
a = input('họ và tên: ')
b = float(input('số điện tháng trước: '))
c = float(input('số điện tháng này: '))
k = c - b
if k <= 50:
    print('số tiền điện là:', k * 1.984)
if k <= 100:
    print('số tiền điện là:', 50 * 1.984 + (k - 50) * 2.050)
if k <= 200:
    print('số tiền điện là:', 50 * 1.984 + 50 * 2.050 + (k - 100) * 2.380)
if k <= 300:
    print('số tiền điện là:', 50 * 1.984 + 50 * 2.050 + 100 * 2.380 + (k - 200) * 2.998)
if k <= 400:
    print('số tiền điện là:', 50 * 1.984 + 50 * 2.050 + 100 * 2.380 + 100 * 2.998 + (k - 300) * 3.350)
if k >= 401:
    print('số tiền điện là:', 50 * 1.984 + 50 * 2.050 + 100 * 2.380 + 100 * 2.998 + 100 * 3.350 + (k - 400) * 3.460)
```
