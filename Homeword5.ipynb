#### W5A16
Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số 
nguyên x 

```python
def amstrong_number(num):
    num_str = str(num)
    num_digits = len(num_str)
    armstrong_sum = sum(int(digit) ** num_digits for digit in num_str)
    return armstrong_sum == num
```

#### W5A18
VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.

```python
def so_doi_xung(n):
    a = n
    b = 0
    while n > 0:
        b = b * 10 + n % 10
        n //= 10
    return a == b
```

#### W5A19
VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.

```python
def so_ky_dieu(n):
    a = 0
    while n > 9:
        a += n % 10
        n //= 10
        if n == 0:
            n = a
    return n == 1
```

#### W5A20
Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.

```python
def detect_number_type(n):
    if type(n).__name__ == 'int':
        return "int"
    elif type(n).__name__ == 'float':
        return "float"
    else:
        return "neither"
```

#### W5A23
Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.

```python
def quadratic_roots_count(a, b, c):
    delta = b ** 2 - 4 * a * c
    if delta > 0:
        return 2
    elif delta == 0:
        return 1
    else:
        return 0
```

Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không. 

```python
def are_collinear(x1, y1, x2, y2, x3, y3):
    return (y2 - y1) * (x3 - x2) == (y3 - y2) * (x2 - x1)
```

#### W5A25
Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay 
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3), 
(0,3)], p = (2,1) 

```python
def point_in_quadrilateral(x1, y1, x2, y2, x3, y3, x4, y4, x, y):
    def area(x1, y1, x2, y2, x3, y3):
        return abs((x1*(y2 - y3) + x2*(y3 - y1) + x3*(y1 - y2)) / 2.0)
    
    A = area(x1, y1, x2, y2, x3, y3) + area(x1, y1, x3, y3, x4, y4) + area(x1, y1, x4, y4, x2, y2) + area(x2, y2, x3, y3, x4, y4)
    A1 = area(x, y, x2, y2, x3, y3)
    A2 = area(x1, y1, x, y, x3, y3)
    A3 = area(x1, y1, x2, y2, x)
    return A == A1 + A2 + A3
```

#### W5A28
Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3). 

```python
def goc(x1, y1, x2, y2, x3, y3):
    import math
    a = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
    b = math.sqrt((x3 - x1) ** 2 + (y3 - y1) ** 2)
    c = math.sqrt((x3 - x2) ** 2 + (y3 - y2) ** 2)
    if a == 0 or b == 0:
        return 0
    cos_theta = (a**2 + b**2 - c**2) / (2 * a * b)
    theta = math.acos(cos_theta)
    return math.degrees(theta)
```

#### W5A30
Viết hàm tính hệ số nhị thức C(n, k)

```python
def Newton_binomial(n, k):
    if k == 0 or k == n:
        return 1
    a = 0
    b = 0
    c = 0
    for i in range(1, k + 1):
        a *= i
    for i in range(1, n - k + 1):
        b *= i  
    for i in range(1, n + 1):
        c *= i
    return c // (a * b)
```

#### W5A33
Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa, 
giá km, phí chờ, phụ phí ban đêm và giảm giá.

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    if car_type == 'standard':
        base_fare = 3
        per_km_fare = 1.5
        wait_fare_per_minute = 0.5
    elif car_type == 'premium':
        base_fare = 5
        per_km_fare = 2.5
        wait_fare_per_minute = 1
    else:
        raise ValueError("Invalid car type")
    
    fare = base_fare + (per_km_fare * km) + (wait_fare_per_minute * wait_minutes)
    
    if 22 <= start_hour < 6:
        fare *= 1.2
    
    return fare
```

#### W5A34
Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau 
Nếu n ≤ 0 trả về 0.

```python
def sum_1_to_n(n):
    if n <= 0:
        return 0
    a = str(n)
    a = set(a)
    a = ''.join(a)
    s = 0 
    if int(a) == n:
        for i in n:
            s += int(i)
        return s
    else:
        return 0
```
